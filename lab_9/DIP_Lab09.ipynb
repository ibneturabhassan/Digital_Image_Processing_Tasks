```
#Task2
import cv2
from skimage.filters.rank import maximum, minimum, geometric_mean
from google.colab.patches import cv2_imshow
import scipy.ndimage.filters as filters
from skimage.util import random_noise
from matplotlib import pyplot as plt 
from skimage.morphology import disk
from skimage.filters import median

img = cv2.imread('pears.png', 0) 
height, width = img.shape
```

```
#Arithematic Mean
noisyimg1 = random_noise(img, 'gaussian')
kernel = np.ones([9,9])/81
filtered = np.zeros([height,width])
filtered = cv2.filter2D(noisyimg1,-1,kernel)
plt.imshow(filtered,cmap="gray")
```

```
#Min_Filter
noisyimg2 = random_noise(img, 'gaussian')
filtered = np.zeros([height,width])
filtered = minimum(noisyimg2,disk(2))
plt.imshow(filtered,cmap="gray")
```

```
#Max Filter
noisyimg = random_noise(img, 'gaussian')
filtered = np.zeros([height,width])
filtered = maximum(noisyimg,disk(1))
plt.imshow(filtered,cmap="gray")
```

```
#Median Filter
noisyimg = random_noise(img, 'gaussian')
filtered = np.zeros([height,width])
filtered = median(noisyimg)
plt.imshow(filtered,cmap="gray")
```

```
#Geometric Mean
img = cv2.imread('pears.png', 0) 
h, w = img.shape
noisyimg = random_noise(img, 'gaussian')
kernel = np.ones([3,3])
filtered = np.zeros([height,width])
filtered = geometric_mean(noisyimg, kernel)
plt.imshow(filtered,cmap="gray")
```

```
#Harmonic Mean
noisyimg = random_noise(img, 'gaussian')
kernel = np.ones([3,3])
filtered = np.zeros([height,width])
filtered = 9/cv2.filter2D(1/noisyimg, -1, kernel)
plt.imshow(filtered,cmap="gray")
```

```
#Contra Harmonic (Positive)
noisyimg = random_noise(img, 'gaussian')
kernel = np.ones([3,3])
q = 3
filtered = np.zeros([height,width])
filtered = cv2.filter2D(noisyimg**(q+1),-1,kernel)/cv2.filter2D(noisyimg**q, -1 ,kernel)
plt.imshow(filtered,cmap="gray")
```
